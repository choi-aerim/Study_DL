```python
# 이미지 resize 사용자 함수 생성
def scale(img_arr, nR, nC):
    n_rows = len(img_arr)    # 행 차원
    n_cols = len(img_arr[0])   # 열 차원
    return [[ img_arr[int(n_rows * r / nR)][int(n_cols * c / nC)]  
                 for c in range(nC)] for r in range(nR)]
```

```python
plt.figure(figsize = (8, 16))
for i in range(4):
    x = plt.imread(f'{train_cats_dir}/cat.{i}.jpg')
    
    plt.subplot(4, 2, i*2 + 1)
    plt.axis('off')
    plt.imshow(x)
    
    x = plt.imread(f'{train_cats_dir}/cat.{i}.jpg')
    x = scale(x, 150, 150)
    plt.subplot(4, 2, i*2 + 2)
    
    plt.axis('off')
    plt.imshow(x)
plt.show()
```

```python
plt.figure(figsize = (8, 16))
for i in range(4):
    x = plt.imread(f'{train_dogs_dir}/dog.{i}.jpg')
    
    plt.subplot(4, 2, i*2 + 1)
    plt.axis('off')
    plt.imshow(x)
    
    x = plt.imread(f'{train_dogs_dir}/dog.{i}.jpg')
    x = scale(x, 150, 150)
    plt.subplot(4, 2, i*2 + 2)
    
    plt.axis('off')
    plt.imshow(x)
plt.show()
```

### 네트워크 구성하기

```python
# 네트워크 구성

from keras import models, layers

model = models.Sequential()
model.add(layers.Conv2D(32, (3,3), activation = 'relu', input_shape = (150,150, 3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation = 'relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128, (3,3), activation = 'relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128, (3,3), activation = 'relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
# 모델 컴파일
from keras import optimizers

model.compile(optimizer = optimizers.RMSprop(learning_rate = 0.0001),
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

## 2. 사전 훈련된 네트워크로 특성 추출 컨브넷 만들기
- 개, 고양이 레이블을 포함한 사전훈련된 네트워크 ImageNet의 vgg16모델을 활용해 보겠음
- 현재 적용하려는 데이터셋 크기가 작고, 사전훈련된 네트워크 데이터셋과 유사한 형태지라, 컨브넷층은 동결하고, 완전연결층 가중치 역시 재사용해도 도움이 될 것 같음
- 하지만, 새로운 문제의 클래스가 원본 모델의 클래스 집합과 겹치지 않는 일반적인 문제를 다루기 위해 특성추출만 진행하고, 완전연결층은 수정해서 사용하겠음

### VGG 네트워크

```python
from keras.applications import VGG16
```

```python
conv_base = VGG16(weights = 'imagenet',
                  include_top = False,
                  input_shape = (150, 150, 3))
```

```python
conv_base.summary()
```

```python
4*4*512
```

- 이 지점에서 <b>두 가지 방식</b>이 가능함
   1. 컨브넷 층 실행하고 출력을 저장한 후, 새로운 완전 연결 분류기에 훈련시키기
       <br>-> pretrain된 모델의 컨브넷 연산은 빠르고 비용이 적게 듦
       하지만, <span style = 'background-color: #ffdce0'> 데이터 증식을 할 수 없음</span><br><br>
       
   2. 준비된 컨브넷 층 모델(conv_base)에 새로운 완전 연결 분류기 층을 추가하여 엔드-투-엔드로 전체 모델을 실행함
      <br> -> 모델에 노출된 모든 입력 이미지가 매번 합성곱 기반 층을 통과하기 때문에 <span style = 'background-color: #ffdce0'>데이터 증식을 사용할 수 있음</span>. 하지만 훨씬 비용이 많이 듬

### 방법1: 컨브넷 층 동결 후 학습/저장 + 새로운 완전 연결분류기 학습

#### 경로 설정

```python
import os
```

```python
# 학습/검증/테스트 디렉토리 경로 설정

base_dir = '../필요데이터/cats_and_dogs_small/'


train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')
```

```python
os.listdir(train_dir), os.listdir(val_dir), os.listdir(test_dir)
```

#### 컨브넷 층 특성 추출

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
# conv_base 출력값을 저장하는 사용자함수 만들기

def extract_features(conv_base, directory, sample_num, batch_size):
    # 출력 특성값 담을 배열 생성
    features = np.zeros((sample_num, 4, 4, 512))
    # 출력 레이블값 담을 배열 생성
    labels = np.zeros((sample_num))
    
    
    # 이미지 제너레이터 객체 만들기
    datagen = ImageDataGenerator(rescale = 1./255)
    img_gen = datagen.flow_from_directory(directory,
                                          target_size = (150,150),
                                          batch_size = batch_size, 
                                          class_mode = 'binary') 
    i = 0
    for input_batch, label_batch in img_gen:
        # 컨브넷베이스 모델 돌리고 출력값 저장
        features_batch = conv_base.predict(input_batch)
        
        features[i * (batch_size): (i+1) * (batch_size)] = features_batch
        labels[i * (batch_size): (i+1) * (batch_size)] = label_batch
        
        i += 1
        if i * batch_size >= sample_num:
            break
    return features, labels
```

```python
train_features, train_labels = extract_features(conv_base, train_dir, 2000, 20)
val_features, val_labels = extract_features(conv_base, val_dir, 1000, 20)
test_features, test_labels = extract_features(conv_base, test_dir, 1000, 20)
```

```python
print(train_features.shape, train_labels.shape)
print(val_features.shape, val_labels.shape)
print(test_features.shape, test_labels.shape)
```

```python
train_features = train_features.reshape(2000, 4*4*512)
val_features = val_features.reshape(1000, 4*4*512)
test_features = test_features.reshape(1000, 4*4*512)
```

```python
print(train_features.shape, train_labels.shape)
print(val_features.shape, val_labels.shape)
print(test_features.shape, test_labels.shape)
```

#### 완전 연결층 생성

```python
from keras import models, layers, optimizers
```

```python
# 완전 연결층 만들기

model = models.Sequential()
model.add(layers.Dense(256, activation = 'relu',input_dim = (4*4*512)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation = 'sigmoid'))

# 모델 컴파일
model.compile(optimizer = optimizers.RMSprop(lr = 2e-5),
              loss = 'binary_crossentropy',
              metrics = ['acc'])
```

#### 완전 연결층 학습

```python
history = model.fit(train_features, train_labels,
                    batch_size = 20,
                    epochs = 30,
                    validation_data = (val_features, val_labels))
```

#### 시각화

```python
import matplotlib.pyplot as plt
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

score = ['acc', 'loss']
val_score = [f'val_{i}' for i in score]

epochs = range(1, 31)

plt.figure(figsize = (12, 5))
for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.plot(epochs, eval(score[i]), 'bo', label = score[i])
    plt.plot(epochs, eval(val_score[i]), 'b', label = val_score[i])
    plt.legend()
plt.show()
```

- 정확도는 90% 내외를 웃돌고 있음

- 큰 비율로 드롭아웃을 시켰음에도 불구하고 6번쯤 만에 과대적합이 시작되고 있음
- 작은데이터셋에도 불구하고 이미지 증식을 시키지 않았기 때문으로 보임

### 방법2: 컨브넷 층  + 새로운 완전 연결분류기에 데이터 증식 후 학습

#### 사전훈련된 컨브넷 + 새로운 완전연결 층 모델 생성

```python
conv_base = VGG16(weights = 'imagenet',
                  include_top = False,
                  input_shape = (150, 150, 3))
```

```python
model = models.Sequential()
model.add(conv_base)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))
```

## 3. 사전 훈련된 네트워크를 미세 튜닝하여 컨브넷 만들기
- 동결한 컨브넷 층을 일부 상위층(완전연결층과 가까운) 몇 개만 제외하고 동결시키는 것
- 새로운 완전연결층도 활용
<br><br>
<b>[순서]</b>
    1. 사전에 훈련된 네트워크 위에 새로운 네트워크 추가
    2. 사전 훈련된 네트워크 동결
    3. 새로 추가한 네트워크 훈련
    4. 사전 훈련된 네트워크에서 일부 층 동결 해제
    5. 동결을 해제한 층과 새로 추가한 층을 함께 훈련
    
- 1,2,3번은 위에서 진행했으므로, 4,5번 실행하기

#### 일부 컨브넷 층 동결 해제

```python
conv_model.summary()
```

- 가장 상위층의 합성곱 3개층을 동결해제 하겠음
<br><br>
- 이유는,
    1. 합성곱 기반 층에 있는 하위 층들은 좀 더 일반적이고 재사용 가능한 특성들을 인코딩함
    2. 훈련 파라미터가 많을수록 과대적합 위험이 큼

- conv_base 모델 컨브넷층은 1500만개의 파라미터를 가지고 있어 작은 데이터로 학습시 과대적합 위험이 큼

```python
print(f'동결 해제 전 훈련 가능한 가중치 수: {len(conv_base.trainable_weights)}')
```

```python
# 동결 해제
conv_base.trainable = True

set_trainable = False
for layer in conv_base.layers:
    if layer.name == 'block5_conv1':
        set_trainable = True
    
    # set_trainable 객체에 따라 모델 층의 훈련 가능성도 바꾸기
    # set_trainable = True로 설정된 층 이후 모든 층은 가중치 훈련이 가능해짐
    if set_trainable:
        layer.trainable = True
    else:
        layer.trainable = False
```

```python
print(f'동결 해제 후 훈련 가능한 가중치 수: {len(conv_base.trainable_weights)}')
```

#### 학습

```python
# 모델 컴파일
model.compile(optimizer = optimizers.RMSprop(lr = 1e-5),
              loss = 'binary_crossentropy',
              metrics = ['acc'])


# 위의 방법2에서 동결된 컨브넷 + 완전연결층에서 훈련한 모델에 이미지 증식 + 상위층 가중치 훈련시키기
history = model.fit(train_gen,
                    steps_per_epoch = 100,
                    epochs = 100,
                    validation_data = val_gen,
                    validation_steps = 50)
```

#### 시각화

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

score = ['acc', 'loss']
val_score = [f'val_{i}' for i in score]

epochs = range(1, 31)

plt.figure(figsize = (12, 5))
for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.plot(epochs, eval(score[i]), 'bo', label = score[i])
    plt.plot(epochs, eval(val_score[i]), 'b', label = val_score[i])
    plt.legend()
plt.show()
```
