```python
drinks = ['pepsi', 'chilsung', 'cocacola', 'milkis','sprite']


for drink in drinks:
  for filename in os.listdir(f'./{drink}'):
    base_dir = f'./{drink}'
    old_filename = os.path.join(base_dir, filename)
    new_filename = os.path.join(base_dir, f'{drink}{filename}')
    os.rename(old_filename, new_filename)
  print(os.listdir(f'./{drink}')[:10])
```
